<h1>Problem Statement</h1>

<p1>Is Singaporeâ€™s University education meeting the demands of the manpower needs in various industries?</p1>

Q1. What are the industries rising in manpower demand?
        - Average monthly recruitment/resignation https://data.gov.sg/dataset/average-monthly-recruitment-resignation-rates-by-industry-and-occupational-group-quarterly
        - Job Vacancy Rate by Industry https://data.gov.sg/dataset/job-vacancy-rates-by-industry-and-occupational-group-quarterly
Q2. Does the employment rates in the Universities correspond with this demand?
        - Graduate Employment Survey https://data.gov.sg/dataset/graduate-employment-survey-ntu-nus-sit-smu-suss-sutd
Q3. Does the Government funding/ University Policies reflect this change?
        - Gov Expenditure on Education https://data.gov.sg/dataset/government-recurrent-expenditure-on-education
        - University Intake by Course https://data.gov.sg/dataset/universities-intake-enrolment-and-graduates-by-course
    

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import re
```

```python
University_IEG = pd.read_csv('universities-intake-enrolment-and-graduates-by-course.csv')
University_IEG['intake'] = University_IEG['intake'].str.replace(",","")
University_IEG['intake'] = pd.to_numeric(University_IEG['intake'])

University_IEG = University_IEG.drop(University_IEG[University_IEG['year']<2013].index)
```

```python
Uni_IEG_pivot = pd.pivot_table(University_IEG, index=['year'], columns=['course'],values='intake',aggfunc=np.sum)
Uni_IEG_pivot
```

```python

Uni_IEG_pivot.plot(kind='bar',stacked=True,figsize=(24,12))
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5),prop={'size': 16})
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.xlabel('Year',fontsize=25)
plt.ylabel('Number of Students',fontsize=25)
plt.title('Intake for University by Year and Field of Study',fontsize=25)
```

```python
Govt_Spending = pd.read_csv('government-recurrent-expenditure-on-education-by-type-of-educational-institutions.csv')
Govt_Spending_student = pd.read_csv('government-recurrent-expenditure-on-education-per-student.csv')
```

```python
Govt_Spending['recurrent_expenditure'] = pd.to_numeric(Govt_Spending['recurrent_expenditure'])

cond1 = Govt_Spending['year']<2013
cond2 = Govt_Spending['type_of_educational_institutions']!='Universities'

Govt_Spending = Govt_Spending.drop(Govt_Spending[Govt_Spending['year']<2013].index)
Govt_Spending = Govt_Spending.drop(Govt_Spending[Govt_Spending['type_of_educational_institutions']!='Universities'].index)

Govt_Spending
```

```python
Govt_Spending.plot(kind='line',figsize=(24,12),x='year', y='recurrent_expenditure')
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.xlabel('Year',fontsize=25)
plt.ylabel('Govt Spending',fontsize=25)
plt.title('Government Recurrent Expenditure for Universities',fontsize=25)
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5),prop={'size': 16})
```

```python
Govt_Spending_student['expenditure_per_student'] = pd.to_numeric(Govt_Spending_student['expenditure_per_student'])


Govt_Spending_student = Govt_Spending_student.drop(Govt_Spending_student[Govt_Spending_student['year']<2013].index)
Govt_Spending_student = Govt_Spending_student.drop(Govt_Spending_student[Govt_Spending_student['type_of_educational_institution']!='Publicly-funded full-time degree courses'].index)

Govt_Spending_student
```

```python
Govt_Spending_df = pd.merge(Govt_Spending,Govt_Spending_student,on='year')
Govt_Spending_df['recurrent_expenditure_in_hundreds'] = Govt_Spending_df['recurrent_expenditure']/100
Govt_Spending_df
```

```python
plt.plot(Govt_Spending_df['year'],Govt_Spending_df['recurrent_expenditure_in_hundreds'],label='Government Recurrent Expenditure On Education')
plt.plot(Govt_Spending_df['year'],Govt_Spending_df['expenditure_per_student'],label='Recurrent amount spent by government per student')
plt.xticks(fontsize=8)
plt.yticks(fontsize=8)
plt.xlabel('Year',fontsize=12)
plt.ylabel('Govt Spending',fontsize=12)
plt.title('Government Recurrent Expenditure for Universities',fontsize=12)
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5),prop={'size': 8})
```
